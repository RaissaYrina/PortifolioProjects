```python
import requests
import json
import pandas as pd

# Inicializa a lista para armazenar todas as revistas
all_magazines = []
page = 1
```

```python
# Testando a conexão com a API
response = requests.get('https://api.openalex.org/sources?filter=country_code:BR&page=1')
print(f'Status Code: {response.status_code}')  # Verifique o código de status

# Verifique o conteúdo retornado
result = response.json()
print(json.dumps(result, indent=2))  # Exibir a resposta formatada
```

```python
# Loop enquanto for verdade
while True:
    # Requisição da API
    response = requests.get(f'https://api.openalex.org/sources?filter=country_code:BR&page={page}')
    
    # Verificando se a solicitação foi bem-sucedida 
    if response.status_code != 200:
        print(f'Erro ao acessar a API: {response.status_code}')
        break
    
    # Ler os dados em formato JSON
    result = response.json()
    magazines = result.get('results', [])
    
    # Se não tiver mais revistas, sair do loop
    if not magazines:
        print(f'Nenhuma revista encontrada na página {page}.')
        break
        
    # Adicionando revistas à lista
    all_magazines.extend(magazines)
    
    # Avance para a próxima página
    page += 1
```

```python
# Verifique quantas revistas foram coletadas
print(f"Número de revistas coletadas: {len(all_magazines)}")

# Se houver revistas, converta para um DataFrame
if all_magazines:
    df = pd.DataFrame(all_magazines)
    print(df.head())  # Mostra as primeiras linhas do DataFrame
else:
    print("Nenhuma revista foi coletada.")
```

```python
# Salve o DataFrame em um arquivo CSV, se houver dados
if not df.empty:
    df.to_csv(r'C:\Users\Raissa Rabello\Desktop\revistas_brasileiras.csv', index=False, encoding='utf-8')
    print("Dados salvos em 'revistas_brasileiras.csv'")
else:
    print("Nada para salvar.")
```
